## After Lunch

groupby, aggregate, method
if you want more than one aggregate, do .agg([put list here])
inside of an agg you can rename the columns if you want more meaningful
if you don't supply aggregating columns, it will do it on all columns

you can define your own aggregating function

in the np.max below, the argumes it takes are from the 'UGDS','SATMTMID'

```python
def mymax(s):
    return s.max() + 1

def max_min(s):
    return s.max() - s.min()

college.groupby(['STABBR', 'HBCU', 'RELAFFIL'])['UGDS','SATMTMID'].agg([np.max, mymax, max_min, lambda s: s.max() - s.min()**2]).head(5)
```

### Filter -- have to have a custom function.  Filtering out records based on the Group

```python
def min_ugds(df):
    return df['UGDS'].sum() > 300000
    
```

```python
college.groupby(['STABBR']).filter(min_ugds).head(3) #returns only states with UGDS greater than 300000
```

### Transform is the oppisite of Aggregation.  like normalize data.  returns same number of records.
changing all the values in the group, pass a custom function
below, we're creating groups based on state and religious affiliation
based on these groups, we're normalizing the population

```python
def mean_normalize(s):
    return (s-s.mean()) / s.std()

g = college.groupby(['STABBR', 'RELAFFIL'])
college['UGDS_norm'] = g['UGDS'].transform(mean_normalize)
college[['STABBR','RELAFFIL', 'UGDS','UGDS_norm']].head(5)
```

```python
def mean_normalize2(s, entire_mean):
    return (s-entire_mean) / s.std()
            
em = college.UGDS.mean()

college['UGDS_all'] = g['UGDS'].transform(mean_normalize2, entire_mean = em)
college[['STABBR','RELAFFIL', 'UGDS','UGDS_norm','UGDS_all']].head(5)
```

```python
#filters out colleges in states that has 100,000 or less

def min_ugds(df, threshold):
    return df['UGDS'].sum() > threshold

college.groupby(['STABBR']).filter(min_ugds, threshold=2000000).head(3) #returns only states with UGDS greater than 300000
```

```python
# create fake data

all_weight = np.zeros(32)
all_weight[::2] = 300 * np.cumprod(np.random.rand(16) * .05 + .96)
all_weight[1::2] = 200 * np.cumprod(np.random.rand(16) * .05 + .96)
all_weight = np.round(all_weight, 0).astype(int)


df_weight = pd.DataFrame({'Name':['Bob', 'Amy'] * 16, 
                          'Month': ('Jan ' * 8 + ' Feb' * 8 + ' Mar' * 8 + ' Apr' * 8).split(),
                          'Week' : (['Week 1'] * 2 + ['Week 2'] * 2 + ['Week 3'] * 2 + ['Week 4'] * 2) * 4,
                          'Weight':all_weight},
                        columns=['Name', 'Month', 'Week', 'Weight'])

df_weight.head(5)
```

```python
def find_perc_loss(s):
    # compare against week one of series
    return (s-s.iloc[0]) / s.iloc[0]

df_weight['pml'] = df_weight.groupby(['Name','Month'])['Weight'].transform(find_perc_loss)
df_weight.head(10)
```

### Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #this sets matplotlib defaults by importing
%matplotlib inline
```

```python
diamonds = pd.read_csv('pandas-learn-plotting/data/diamonds.csv')
sns.countplot('cut', data=diamonds)
```

```python
sns.countplot('cut', hue='color', data=diamonds)
```

```python
sns.distplot(diamonds.price, kde=False)
```

```python
sns.jointplot('carat','price', data=diamonds)
```

```python
sns.lmplot('carat',
           'price',
           data=diamonds.sample(frac=.01),
           hue='cut', col='clarity',
           #col_wrap = 3,
           row='color',
           sharex=True,
           sharey=False,
           fit_reg=False)
#row, column, hue, carat price as dimensions
```

```python
sns.boxplot('cut','price',data=diamonds,hue='color')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
sns.boxplot('cut','price',data=diamonds,hue='color', ax=ax2)
```
