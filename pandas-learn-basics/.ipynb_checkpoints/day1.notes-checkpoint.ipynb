```python
from pandas_datareader import data
```

```python
tqqq = data.DataReader('TQQQ','google')
```

```python
type(tqqq)
```

```python
close = tqqq['Close']
```

```python
close.iloc[-1]
```

```python
close.index
```

```python
close.iloc[[3,5,7,8,90]] # pass a list
           
```

```python
dpc = close.pct_change().dropna() #daily percent change
```

```python
dpc.head()
```

```python
%matplotlib inline
```

```python
dpc.hist() #is this normally distriburted?
```

```python
std = dpc.std()
mean = dpc.mean()
criteria_1 = (dpc <= (mean+1*std)) & (dpc >= (mean-1*std)) #must use amperstand
criteria_2 = (dpc <= (mean+2*std)) & (dpc >= (mean-2*std)) #must use amperstand
criteria_3 = (dpc <= (mean+3*std)) & (dpc >= (mean-3*std)) #must use amperstand
```

```python
criteria_1.sum()/criteria_1.count() # if this is around 68% then it's normal
```

```python
criteria_1.sum()/criteria_1.count(), criteria_2.sum()/criteria_2.count(),criteria_3.sum()/criteria_3.count()
```
