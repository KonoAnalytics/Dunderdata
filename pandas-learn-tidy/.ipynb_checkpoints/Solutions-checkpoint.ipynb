### Problem 4
<span  style="color:green; font-size:16px">Find a different variable in the columns and tidy that variable by creating another three column DataFrame</span>

```python
melted_cols = ['COUNT PACIFIC ISLANDER', 
               'COUNT HISPANIC LATINO', 
               'COUNT AMERICAN INDIAN', 
               'COUNT ASIAN NON HISPANIC', 
               'COUNT WHITE NON HISPANIC',
               'COUNT BLACK NON HISPANIC',
               'COUNT OTHER ETHNICITY',
               'COUNT ETHNICITY UNKNOWN']

df_count = pd.melt(ny_demo, 
             id_vars='JURISDICTION NAME', 
             value_vars=melted_cols,
             var_name='Ethnicity', 
             value_name='Count')

# df_count = df_count.query('Count > 0')

df_count['Ethnicity'] = df_count.Ethnicity.str.replace('COUNT ', '')
df_count.head(10)
```

### Problem 5
<span  style="color:green; font-size:16px">For the same variable you used in problem 4, create another three column tidy dataset using the percentage column instead of the count.</span>

```python
melted_cols = ['PERCENT PACIFIC ISLANDER', 
               'PERCENT HISPANIC LATINO', 
               'PERCENT AMERICAN INDIAN', 
               'PERCENT ASIAN NON HISPANIC', 
               'PERCENT WHITE NON HISPANIC',
               'PERCENT BLACK NON HISPANIC',
               'PERCENT OTHER ETHNICITY',
               'PERCENT ETHNICITY UNKNOWN']

df_perc = pd.melt(ny_demo, 
             id_vars='JURISDICTION NAME', 
             value_vars=melted_cols,
             var_name='Ethnicity', 
             value_name='Percent')

# df_perc = df_perc.query('Percent > 0')

df_perc['Ethnicity'] = df_perc.Ethnicity.str.replace('PERCENT ', '')
df_perc.head(10)
```

```python
def greatest_common_factor(x, y):
    '''
    Find the greatest common factor between x and y

    Parameters
    ----------
    x: a positive integer
    y: a positive integer

    Returns
    -------
    The largest integer divisible by x and y
    '''
    x_factors = [num for num in range(1, x+1) if x % num == 0]
    y_factors = [num for num in range(1, y+1) if y % num == 0]
    return max(set(x_factors) & set(y_factors))
```

### Problem 10
<span  style="color:green; font-size:16px">Use the `melt` function to make make the DataFrame `coh_pivot` tidy.</span>

```python
coh_pivot = coh.pivot_table(index=['RACE', 'DEPARTMENT'], 
                            columns='GENDER', 
                            values='BASE_SALARY', 
                            aggfunc=np.mean)\
                .reset_index()\
                .rename_axis(None, axis='columns')
coh_pivot.head(10)
```

```python
pd.melt(coh_pivot, 
        id_vars=['RACE', 'DEPARTMENT'], 
        value_vars=['Female', 'Male'], 
        value_name='GENDER',
        var_name='BASE_SALARY')\
  .head(15)
```

### Problem 11
<span  style="color:green; font-size:16px">Use the `stack` function to make make the DataFrame `coh_pivot` from problem 10 tidy.</span>

```python
coh_pivot.set_index(['RACE', 'DEPARTMENT'])\
         .stack()\
         .rename_axis(['RACE', 'DEPARTMENT', 'GENDER'])\
         .rename('BASE_SALARY')\
         .reset_index()\
         .head(10)
```

### Problem 2
<span  style="color:green; font-size:16px">Tidy the dataset Impaired_Driving_Death_Rate.csv</span>

```python
df = pd.read_csv('data/Impaired_Driving_Death_Rate.csv')
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
lat_long = df.Location.str.extract('(\d+\.\d+, \-\d+\.\d+)', expand=False)
lat_long.head()
```

```python
df[['lat', 'long']] = lat_long.str.split(', ', expand=True).astype(float)
df = df.drop('Location', axis=1)
```

```python
df.head(10)
```

```python
df_melt = pd.melt(df, id_vars=['State', 'lat', 'long'], value_name='Rate')
```

```python
df_melt.variable.value_counts()
```

```python
gender = np.where(df_melt.variable.str.contains('Male'), 'Male', 
        np.where(df_melt.variable.str.contains('Female'), 'Female', 'All'))
```

```python
year = df_melt.variable.str[-4:].astype(int)
```

```python
year.head()
```

```python
age = df_melt.variable.str.extract('(\d?\d+\-\\d\d,)', expand=False).fillna('All')
```

```python
df_melt
```

```python
df_melt['gender'] = gender
df_melt['year'] = year
df_melt['age'] = age
```

```python
df_melt = df_melt.drop('variable', axis=1)
```
